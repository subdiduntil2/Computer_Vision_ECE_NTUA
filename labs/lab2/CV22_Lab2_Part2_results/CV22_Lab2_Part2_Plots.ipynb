```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from numpy import load
```

```python
# load array
H_Harris_walk = load('H_Harris_walk.npy')
H_Gabor_walk = load('H_Gabor_walk.npy')
H_Harris_box = load('H_Harris_box.npy')
H_Gabor_box = load('H_Gabor_box.npy')
H_Harris_run = load('H_Harris_run.npy')
H_Gabor_run = load('H_Gabor_run.npy')
```

```python
fig, ax = plt.subplots(1, 3, figsize = (20,5))
ax[0].imshow(H_Harris_walk[:,:,50], cmap='gray_r')
ax[2].imshow(H_Harris_walk[:,:,51], cmap='gray_r')
ax[1].imshow(H_Harris_walk[:,:,52], cmap='gray_r')
ax[0].set_title("Frame 50", fontsize=15);
ax[2].set_title("Frame 51", fontsize=15);
ax[1].set_title("Frame 52", fontsize=15);
plt.suptitle("   - walk video 1 - Harris Detector", fontsize=20);

fig, ax = plt.subplots(1, 3, figsize = (20,5))
ax[0].imshow(H_Gabor_walk[:,:,50], cmap='gray_r')
ax[2].imshow(H_Gabor_walk[:,:,51], cmap='gray_r')
ax[1].imshow(H_Gabor_walk[:,:,52], cmap='gray_r')
ax[0].set_title("Frame 50", fontsize=15);
ax[2].set_title("Frame 51", fontsize=15);
ax[1].set_title("Frame 52", fontsize=15);
plt.suptitle("   - walk video 1 - Gabor Detector", fontsize=20);
```
